```python
import numpy as np
```

```python
# Create an Array of evenly spaced items from 10 to 20,
a = np.linspace(10,20) # it creates an array of  50 numbers from 10 to 20
a
```

```python
# The above is same as
a = np.linspace(10,20,num=50) 
a
```

```python
a = np.linspace(10,50,num=20) # it creates an array of 20 numbers from 10 to 50
a
```

```python
# Create 100 numbers from 10 to 20 but not included 20
a = np.linspace(10,20,num=100,endpoint=False) 
a
```

```python
# Create 50 numbers from 10 to 20 but not included 20(stop value) and step value is 0.2
a = np.linspace(10,20,num=50,endpoint=False, retstep=True) 
a
```
